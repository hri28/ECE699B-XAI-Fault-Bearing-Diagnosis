```python
import numpy as np, cv2, matplotlib.pyplot as plt

def gradcam(model, img_40x40x1, class_index=None, conv_layer_name=None):
    if conv_layer_name is None:
        conv_layer_name = [l.name for l in model.layers if "conv" in l.name][-1]
    conv_layer = model.get_layer(conv_layer_name)
    x = np.expand_dims(img_40x40x1, 0).astype(np.float32)

    with tf.GradientTape() as tape:
        grad_model = tf.keras.Model(inputs=model.input,
                                    outputs=[conv_layer.output, model.output])
        conv_out, preds = grad_model(x, training=False)
        if class_index is None:
            class_index = int(tf.argmax(preds[0]))
        loss = preds[:, class_index]

    grads = tape.gradient(loss, conv_out)
    weights = tf.reduce_mean(grads, axis=(0,1,2)).numpy()
    conv_out = conv_out[0].numpy()

    cam = np.tensordot(conv_out, weights, axes=([2],[0]))
    cam = np.maximum(cam, 0)
    cam = cv2.resize(cam, (img_40x40x1.shape[1], img_40x40x1.shape[0]))
    cam /= cam.max() + 1e-8
    return cam, int(class_index)

def overlay(cam, img, alpha=0.5):
    img_rgb = np.stack([img]*3, axis=-1)
    heatmap = cv2.applyColorMap((cam*255).astype(np.uint8), cv2.COLORMAP_JET)
    heatmap = cv2.cvtColor(heatmap, cv2.COLOR_BGR2RGB)/255.
    return np.clip((1-alpha)*img_rgb + alpha*heatmap, 0, 1)
```

```python
"""
# -----------------------------------------------------------------------------
# Fixed Grad-CAM Integration - Add this at the end of your existing script
# -----------------------------------------------------------------------------
"""

# Import additional required libraries
import cv2

# Define fault names for better visualization
fault_names = [
    'Normal', 'Inner Race 0.18mm', 'Ball 0.18mm', 'Outer Race 0.18mm',
    'Inner Race 0.36mm', 'Ball 0.36mm', 'Outer Race 0.36mm',
    'Inner Race 0.54mm', 'Ball 0.54mm', 'Outer Race 0.54mm'
]

# Fixed Grad-CAM implementation
def compute_gradcam_fixed(model, img_array, layer_name, pred_class=None):
    """Compute Grad-CAM heatmap for a given image and layer"""

    # Ensure input has batch dimension
    if len(img_array.shape) == 3:
        img_array = img_array[np.newaxis, ...]

    # Create a new model that outputs both the conv layer and final output
    # Get the layer object
    conv_layer = None
    for layer in model.layers:
        if layer.name == layer_name:
            conv_layer = layer
            break

    if conv_layer is None:
        print(f"Layer {layer_name} not found!")
        return None

    # Build a model that maps inputs to conv outputs and predictions
    grad_model = tf.keras.Model(
        inputs=model.inputs,
        outputs=[conv_layer.output, model.output]
    )

    # Cast image to float32
    img_array = tf.cast(img_array, tf.float32)

    # Record operations for automatic differentiation
    with tf.GradientTape() as tape:
        conv_outputs, predictions = grad_model(img_array)
        if pred_class is None:
            pred_class = tf.argmax(predictions[0])
        class_channel = predictions[:, pred_class]

    # Compute gradients
    grads = tape.gradient(class_channel, conv_outputs)

    # Pool gradients and weight conv outputs
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))
    conv_outputs = conv_outputs[0]
    pooled_grads = pooled_grads.numpy()
    conv_outputs = conv_outputs.numpy()

    # Weight the channels by corresponding gradients
    for i in range(pooled_grads.shape[-1]):
        conv_outputs[:, :, i] *= pooled_grads[i]

    # Create heatmap
    heatmap = np.mean(conv_outputs, axis=-1)

    # Normalize
    heatmap = np.maximum(heatmap, 0)
    heatmap = heatmap / np.max(heatmap) if np.max(heatmap) != 0 else heatmap

    return heatmap

# Alternative simpler approach using output gradients
def simple_gradient_visualization(model, img_array, class_idx=None):
    """Simple gradient-based visualization"""
    if len(img_array.shape) == 3:
        img_array = img_array[np.newaxis, ...]

    img_tensor = tf.convert_to_tensor(img_array, dtype=tf.float32)

    with tf.GradientTape() as tape:
        tape.watch(img_tensor)
        predictions = model(img_tensor)

        if class_idx is None:
            class_idx = tf.argmax(predictions[0])

        target_score = predictions[0, class_idx]

    gradients = tape.gradient(target_score, img_tensor)
    gradients = tf.abs(gradients)
    gradients = tf.reduce_max(gradients, axis=-1)

    return gradients.numpy().squeeze()

# Quick visualization
print("\n=== Running Grad-CAM Analysis ===")

# First, let's check available layers
print("\nAvailable convolutional layers in the model:")
conv_layers = []
for i, layer in enumerate(CNN_2D_best_model.layers):
    print(f"Layer {i}: {layer.name} - {type(layer).__name__}")
    if 'Conv2D' in str(type(layer)):
        conv_layers.append(layer.name)

if not conv_layers:
    # If model is Sequential, we need to access layers differently
    for i, layer in enumerate(CNN_2D_best_model.layers[0].layers if hasattr(CNN_2D_best_model.layers[0], 'layers') else CNN_2D_best_model.layers):
        if 'Conv2D' in str(type(layer)):
            conv_layers.append(f"conv2d_{len(conv_layers)}" if 'conv2d' not in layer.name else layer.name)

print(f"\nFound convolutional layers: {conv_layers}")

# Select samples to visualize
num_samples = min(10, len(X_2D_test))
indices = np.random.choice(len(X_2D_test), num_samples, replace=False)

# Create figure with gradient visualization (more reliable)
fig, axes = plt.subplots(num_samples, 3, figsize=(12, 4*num_samples))
if num_samples == 1:
    axes = axes.reshape(1, -1)

print("\nGenerating visualizations...")

for i, idx in enumerate(indices):
    # Get image and labels
    img = X_2D_test[idx]
    true_label = np.argmax(y_2D_test[idx])

    # Get prediction
    pred = CNN_2D_best_model.predict(img[np.newaxis, ...], verbose=0)
    pred_label = np.argmax(pred[0])
    pred_conf = pred[0][pred_label] * 100

    # Use simple gradient visualization
    gradient_map = simple_gradient_visualization(CNN_2D_best_model, img, pred_label)

    # Plot original image
    axes[i, 0].imshow(img.squeeze(), cmap='gray')
    axes[i, 0].set_title(f'Input Signal\nTrue: {fault_names[true_label]}', fontsize=10)
    axes[i, 0].axis('off')

    # Plot gradient map
    im = axes[i, 1].imshow(gradient_map, cmap='hot')
    axes[i, 1].set_title(f'Important Regions\nPred: {fault_names[pred_label]} ({pred_conf:.1f}%)', fontsize=10)
    axes[i, 1].axis('off')

    # Overlay
    axes[i, 2].imshow(img.squeeze(), cmap='gray')
    axes[i, 2].imshow(gradient_map, cmap='hot', alpha=0.5)
    correct = '✓' if pred_label == true_label else '✗'
    axes[i, 2].set_title(f'Overlay {correct}', fontsize=10)
    axes[i, 2].axis('off')

plt.suptitle('CNN Decision Visualization - Gradient Analysis', fontsize=14)
plt.tight_layout()

# Save the figure
save_path = foldername + 'gradient_visualization.png'
plt.savefig(save_path, dpi=300, bbox_inches='tight')
print(f"\nSaved visualization to: {save_path}")
plt.show()

# Try Grad-CAM with proper layer access
if len(conv_layers) > 0:
    print("\nAttempting Grad-CAM visualization...")

    # Use the last conv layer
    target_layer_idx = -1  # Last conv layer usually gives best results

    fig2, axes2 = plt.subplots(min(5, num_samples), 4, figsize=(16, 4*min(5, num_samples)))
    if min(5, num_samples) == 1:
        axes2 = axes2.reshape(1, -1)

    for i in range(min(5, num_samples)):
        idx = indices[i]
        img = X_2D_test[idx]
        true_label = np.argmax(y_2D_test[idx])

        pred = CNN_2D_best_model.predict(img[np.newaxis, ...], verbose=0)
        pred_label = np.argmax(pred[0])
        pred_conf = pred[0][pred_label] * 100

        # Try to compute Grad-CAM
        try:
            # Access the conv layer directly from the model
            if hasattr(CNN_2D_best_model, 'layers'):
                # For Sequential model, we need to access the actual conv layers
                all_layers = []
                for layer in CNN_2D_best_model.layers:
                    if hasattr(layer, 'layers'):  # If it's a container
                        all_layers.extend(layer.layers)
                    else:
                        all_layers.append(layer)

                # Find conv layers
                conv_layer_objects = [l for l in all_layers if 'Conv2D' in str(type(l))]

                if conv_layer_objects:
                    target_layer = conv_layer_objects[target_layer_idx]

                    # Create grad model
                    grad_model = tf.keras.Model(
                        inputs=CNN_2D_best_model.inputs,
                        outputs=[target_layer.output, CNN_2D_best_model.output]
                    )

                    # Compute heatmap
                    img_array = tf.cast(img[np.newaxis, ...], tf.float32)

                    with tf.GradientTape() as tape:
                        conv_outputs, predictions = grad_model(img_array)
                        class_channel = predictions[:, pred_label]

                    grads = tape.gradient(class_channel, conv_outputs)
                    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))

                    conv_outputs = conv_outputs[0].numpy()
                    pooled_grads = pooled_grads.numpy()

                    for j in range(pooled_grads.shape[-1]):
                        conv_outputs[:, :, j] *= pooled_grads[j]

                    heatmap = np.mean(conv_outputs, axis=-1)
                    heatmap = np.maximum(heatmap, 0)
                    heatmap = heatmap / np.max(heatmap) if np.max(heatmap) != 0 else heatmap

                    # Resize heatmap
                    heatmap_resized = cv2.resize(heatmap, (40, 40))

                    # Plot
                    axes2[i, 0].imshow(img.squeeze(), cmap='gray')
                    axes2[i, 0].set_title(f'Original\nTrue: {fault_names[true_label]}', fontsize=10)
                    axes2[i, 0].axis('off')

                    axes2[i, 1].imshow(heatmap_resized, cmap='jet')
                    axes2[i, 1].set_title(f'Grad-CAM\nPred: {fault_names[pred_label]}', fontsize=10)
                    axes2[i, 1].axis('off')

                    axes2[i, 2].imshow(img.squeeze(), cmap='gray')
                    axes2[i, 2].imshow(heatmap_resized, cmap='jet', alpha=0.5)
                    axes2[i, 2].set_title(f'Overlay', fontsize=10)
                    axes2[i, 2].axis('off')

                    # Important regions
                    threshold = np.percentile(heatmap_resized, 75)
                    mask = heatmap_resized > threshold
                    axes2[i, 3].imshow(img.squeeze() * mask, cmap='gray')
                    axes2[i, 3].set_title('Key Regions', fontsize=10)
                    axes2[i, 3].axis('off')

        except Exception as e:
            print(f"Could not compute Grad-CAM for sample {i}: {str(e)}")
            # Fall back to gradient visualization
            gradient_map = simple_gradient_visualization(CNN_2D_best_model, img, pred_label)
            for j in range(4):
                axes2[i, j].imshow(gradient_map if j > 0 else img.squeeze(),
                                  cmap='hot' if j > 0 else 'gray')
                axes2[i, j].axis('off')

    plt.suptitle('Grad-CAM Visualization', fontsize=14)
    plt.tight_layout()
    save_path = foldername + 'gradcam_visualization_attempted.png'
    plt.savefig(save_path, dpi=300, bbox_inches='tight')
    plt.show()

print("\n=== Visualization Complete ===")

# Analyze patterns by class
print("\nAnalyzing patterns by fault class...")

class_gradients = {i: [] for i in range(10)}
predictions = CNN_2D_best_model.predict(X_2D_test, verbose=0)
pred_classes = np.argmax(predictions, axis=1)
true_classes = np.argmax(y_2D_test, axis=1)

# Collect gradients for correctly classified samples
for i in range(min(len(X_2D_test), 200)):  # Limit to 200 samples for speed
    if pred_classes[i] == true_classes[i]:
        grad_map = simple_gradient_visualization(CNN_2D_best_model, X_2D_test[i], true_classes[i])
        class_gradients[true_classes[i]].append(grad_map)

# Plot average patterns
fig3, axes3 = plt.subplots(2, 5, figsize=(20, 8))
axes3 = axes3.ravel()

for class_idx in range(10):
    if len(class_gradients[class_idx]) > 0:
        avg_pattern = np.mean(class_gradients[class_idx], axis=0)
    else:
        avg_pattern = np.zeros((40, 40))

    im = axes3[class_idx].imshow(avg_pattern, cmap='hot')
    axes3[class_idx].set_title(f'{fault_names[class_idx]}\n({len(class_gradients[class_idx])} samples)',
                               fontsize=10)
    axes3[class_idx].axis('off')
    plt.colorbar(im, ax=axes3[class_idx], fraction=0.046, pad=0.04)

plt.suptitle('Average Important Regions by Fault Type', fontsize=14)
plt.tight_layout()

save_path = foldername + 'average_importance_patterns.png'
plt.savefig(save_path, dpi=300, bbox_inches='tight')
print(f"Saved average patterns to: {save_path}")
plt.show()

print("\n=== Analysis Complete ===")
```

```python
class LRP_Explainer:
    """
    Layer-wise Relevance Propagation (LRP) with epsilon rule
    Reference: "Layer-Wise Relevance Propagation: An Overview" (Montavon et al., 2019)
    """

    def __init__(self, model, epsilon=1e-4):
        """
        Args:
            model: Trained Keras model
            epsilon: Small value for numerical stability (epsilon-rule)
        """
        self.model = model
        self.epsilon = epsilon

    def lrp_dense(self, R, layer, A_prev, A):
        """
        LRP for Dense layer (epsilon-rule)
        R: Relevance from next layer
        A_prev: Activations from previous layer
        A: Activations from current layer
        """
        W = layer.get_weights()[0]  # Weights
        B = layer.get_weights()[1] if len(layer.get_weights()) > 1 else 0  # Bias

        # Forward pass: z = W * A_prev + B
        z = np.dot(A_prev, W) + B + self.epsilon * np.sign(A)

        # Backward pass: R_prev = A_prev * (W * R / z)
        s = R / z
        c = np.dot(s, W.T)
        R_prev = A_prev * c

        return R_prev

    def lrp_conv2d(self, R, layer, A_prev, A):
        """
        LRP for Conv2D layer (epsilon-rule)
        """
        # Get layer weights
        W = layer.get_weights()[0]  # Conv filters
        B = layer.get_weights()[1] if len(layer.get_weights()) > 1 else 0

        # Get layer config
        strides = layer.strides
        padding = layer.padding

        # Forward pass to get z
        z = tf.nn.conv2d(
            A_prev,
            W,
            strides=strides,
            padding=padding.upper()
        ).numpy() + B + self.epsilon * np.sign(A)

        # Backward pass
        s = R / z

        # Transpose convolution (deconvolution)
        c = tf.nn.conv2d_transpose(
            s,
            W,
            output_shape=A_prev.shape,
            strides=strides,
            padding=padding.upper()
        ).numpy()

        R_prev = A_prev * c

        return R_prev

    def lrp_maxpool(self, R, layer, A_prev, A):
        """
        LRP for MaxPooling2D layer
        Distributes relevance only to the winning (max) positions
        """
        pool_size = layer.pool_size
        strides = layer.strides
        padding = layer.padding

        # Get the shape
        batch_size, height, width, channels = A_prev.shape

        # Create output to store relevances
        R_prev = np.zeros_like(A_prev)

        # For each position in the pooled output
        for i in range(A.shape[1]):  # height
            for j in range(A.shape[2]):  # width
                # Calculate the window in the input
                h_start = i * strides[0]
                h_end = h_start + pool_size[0]
                w_start = j * strides[1]
                w_end = w_start + pool_size[1]

                # Get the window
                window = A_prev[0, h_start:h_end, w_start:w_end, :]

                # Find max positions
                for c in range(channels):
                    max_val = A[0, i, j, c]
                    mask = (window[:, :, c] == max_val)

                    # Distribute relevance to max positions only
                    R_prev[0, h_start:h_end, w_start:w_end, c] += (
                        R[0, i, j, c] * mask / (mask.sum() + 1e-8)
                    )

        return R_prev

    def explain(self, x, class_index=None):
        """
        Compute LRP heatmap for input x

        Args:
            x: Input image (H, W, C)
            class_index: Target class (None = predicted class)

        Returns:
            Relevance heatmap (H, W)
        """
        # Expand dims for batch
        x_batch = np.expand_dims(x, axis=0).astype(np.float32)

        # Forward pass: collect activations
        activations = [x_batch]
        current = x_batch

        for layer in self.model.layers:
            current = layer(current)
            activations.append(current.numpy())

        # Get prediction
        preds = activations[-1]
        if class_index is None:
            class_index = int(np.argmax(preds[0]))

        # Initialize relevance at output layer
        R = np.zeros_like(preds)
        R[0, class_index] = preds[0, class_index]

        # Backward pass: propagate relevances
        for i in range(len(self.model.layers) - 1, -1, -1):
            layer = self.model.layers[i]
            A = activations[i + 1]  # Current layer output
            A_prev = activations[i]  # Previous layer output (current input)

            layer_type = layer.__class__.__name__

            if layer_type == 'Dense':
                # Flatten if needed
                if len(R.shape) > 2:
                    R = R.reshape(R.shape[0], -1)
                if len(A_prev.shape) > 2:
                    A_prev = A_prev.reshape(A_prev.shape[0], -1)

                R = self.lrp_dense(R, layer, A_prev, A)

            elif layer_type == 'Conv2D':
                R = self.lrp_conv2d(R, layer, A_prev, A)

            elif layer_type == 'MaxPooling2D':
                R = self.lrp_maxpool(R, layer, A_prev, A)

            elif layer_type == 'Flatten':
                # Reshape back to conv shape
                R = R.reshape(A_prev.shape)

            elif layer_type in ['Dropout', 'BatchNormalization']:
                # Pass through
                continue

            elif 'Activation' in layer_type or layer_type == 'ReLU':
                # Pass through activation
                continue

        # Extract heatmap
        heatmap = R[0].sum(axis=-1) if len(R.shape) > 3 else R[0]

        # Normalize
        heatmap = np.abs(heatmap)  # Take absolute values
        heatmap = (heatmap - heatmap.min()) / (heatmap.max() - heatmap.min() + 1e-8)

        return heatmap, class_index


def compute_lrp(model, img_array, class_index=None, epsilon=1e-4):
    """
    Wrapper function for LRP computation
    """
    explainer = LRP_Explainer(model, epsilon=epsilon)
    heatmap, pred_class = explainer.explain(img_array, class_index)
    return heatmap, pred_class


# ============================================================================
# 3. USAGE EXAMPLES
# ============================================================================

def example_usage(model, X_test, y_test, save_dir="explainability_results"):
    """
    Example of how to use both methods
    """
    os.makedirs(save_dir, exist_ok=True)

    # Select a test sample
    idx = 0
    img = X_test[idx]
    true_class = int(np.argmax(y_test[idx]))

    # ========== GRAD-CAM ==========
    print("Computing Grad-CAM...")

    # Find the last convolutional layer
    conv_layers = [l.name for l in model.layers if 'conv' in l.name]
    last_conv = conv_layers[-1] if conv_layers else None

    if last_conv:
        gradcam_heat, pred_class = gradcam_fixed(
            model, img,
            class_index=true_class,
            conv_layer_name=last_conv
        )

        # Visualize Grad-CAM
        fig, axes = plt.subplots(1, 3, figsize=(12, 4))

        axes[0].imshow(img.squeeze(), cmap='gray')
        axes[0].set_title('Input Image')
        axes[0].axis('off')

        axes[1].imshow(gradcam_heat, cmap='jet')
        axes[1].set_title('Grad-CAM Heatmap')
        axes[1].axis('off')

        axes[2].imshow(overlay_heatmap(gradcam_heat, img.squeeze()))
        axes[2].set_title(f'Overlay (True={true_class}, Pred={pred_class})')
        axes[2].axis('off')

        plt.tight_layout()
        plt.savefig(f"{save_dir}/gradcam_example_{idx}.png", dpi=150, bbox_inches='tight')
        plt.show()

    # ========== LRP ==========
    print("Computing LRP...")

    lrp_heat, pred_class = compute_lrp(
        model, img,
        class_index=true_class,
        epsilon=1e-4
    )

    # Visualize LRP
    fig, axes = plt.subplots(1, 3, figsize=(12, 4))

    axes[0].imshow(img.squeeze(), cmap='gray')
    axes[0].set_title('Input Image')
    axes[0].axis('off')

    axes[1].imshow(lrp_heat, cmap='jet')
    axes[1].set_title('LRP Heatmap')
    axes[1].axis('off')

    # Create overlay for LRP
    img_rgb = np.stack([img.squeeze()] * 3, axis=-1)
    heat_colored = cv2.applyColorMap((lrp_heat * 255).astype(np.uint8), cv2.COLORMAP_JET)
    heat_colored = cv2.cvtColor(heat_colored, cv2.COLOR_BGR2RGB) / 255.0
    lrp_overlay = np.clip(0.5 * img_rgb + 0.5 * heat_colored, 0, 1)

    axes[2].imshow(lrp_overlay)
    axes[2].set_title(f'Overlay (True={true_class}, Pred={pred_class})')
    axes[2].axis('off')

    plt.tight_layout()
    plt.savefig(f"{save_dir}/lrp_example_{idx}.png", dpi=150, bbox_inches='tight')
    plt.show()

    print(f"Visualizations saved to {save_dir}/")
```

```python
# ============================================================================
# ULTRA-SIMPLE GRAD-CAM - DIRECT GRADIENT APPROACH
# No model reconstruction, just pure gradient computation
# ============================================================================

import tensorflow as tf
import numpy as np
import cv2
import matplotlib.pyplot as plt
import os

print("=" * 70)
print("ULTRA-SIMPLE GRAD-CAM")
print("=" * 70)

# ============================================================================
# CRITICAL: Call model once to "build" it
# ============================================================================
print("\n🔧 Building model by calling it once...")
_ = CNN_2D_best_model(X_2D_test[:1])
print("✓ Model built!")

# ============================================================================
# Find last conv layer
# ============================================================================
print("\n🔍 Finding last convolutional layer...")
last_conv_layer = None
for layer in reversed(CNN_2D_best_model.layers):
    if 'Conv2D' in str(type(layer).__name__):
        last_conv_layer = layer
        print(f"✓ Found: {layer.name}")
        break

if last_conv_layer is None:
    raise ValueError("No Conv2D layer found!")

# ============================================================================
# Simple Grad-CAM function using direct approach
# ============================================================================

def compute_gradcam_simple(model, image, class_idx, conv_layer):
    """
    Simplest possible Grad-CAM implementation
    """
    # Add batch dimension
    if len(image.shape) == 3:
        image = np.expand_dims(image, axis=0)

    image = tf.cast(image, tf.float32)

    # Watch the conv layer output
    with tf.GradientTape() as tape:
        # Forward pass - get conv output and predictions
        conv_output = None
        x = image

        # Pass through layers until we reach the target conv layer
        for layer in model.layers:
            x = layer(x)
            if layer == conv_layer:
                conv_output = x
                tape.watch(conv_output)

        # x is now the final output
        predictions = x

        # Get the score for target class
        if class_idx is None:
            class_idx = tf.argmax(predictions[0])
        target_class_output = predictions[:, class_idx]

    # Compute gradients
    grads = tape.gradient(target_class_output, conv_output)

    # Global average pooling on gradients
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))

    # Weight the channels by the gradients
    conv_output = conv_output.numpy()[0]
    pooled_grads = pooled_grads.numpy()

    for i in range(len(pooled_grads)):
        conv_output[:, :, i] *= pooled_grads[i]

    # Create heatmap
    heatmap = np.mean(conv_output, axis=-1)
    heatmap = np.maximum(heatmap, 0)

    if heatmap.max() > 0:
        heatmap = heatmap / heatmap.max()

    return heatmap, int(class_idx)


def create_overlay(heatmap, image, alpha=0.5):
    """Create colored overlay"""
    # Resize heatmap to match image size
    heatmap_resized = cv2.resize(heatmap, (image.shape[1], image.shape[0]))

    # Convert to RGB
    if len(image.shape) == 2:
        image = np.stack([image] * 3, axis=-1)
    elif image.shape[-1] == 1:
        image = np.repeat(image, 3, axis=-1)

    if image.max() > 1:
        image = image / 255.0

    # Apply colormap
    heatmap_colored = cv2.applyColorMap(
        (heatmap_resized * 255).astype(np.uint8),
        cv2.COLORMAP_JET
    )
    heatmap_colored = cv2.cvtColor(heatmap_colored, cv2.COLOR_BGR2RGB) / 255.0

    # Blend
    overlay = alpha * heatmap_colored + (1 - alpha) * image

    return np.clip(overlay, 0, 1), heatmap_resized


# ============================================================================
# Create output directory
# ============================================================================
gradcam_dir = f'{foldername}gradcam_outputs/'
os.makedirs(gradcam_dir, exist_ok=True)
print(f"\nOutput directory: {gradcam_dir}")

# ============================================================================
# Generate visualizations
# ============================================================================

print("\n" + "=" * 70)
print("GENERATING GRAD-CAM VISUALIZATIONS")
print("=" * 70)

num_samples = min(10, len(X_2D_test))
indices = np.random.choice(len(X_2D_test), num_samples, replace=False)

fig, axes = plt.subplots(num_samples, 3, figsize=(12, 4*num_samples))
if num_samples == 1:
    axes = axes.reshape(1, -1)

print(f"\nGenerating Grad-CAM for {num_samples} samples...\n")

success_count = 0
for i, idx in enumerate(indices):
    img = X_2D_test[idx]
    true_label = np.argmax(y_2D_test[idx])

    # Get prediction
    pred = CNN_2D_best_model.predict(img[np.newaxis, ...], verbose=0)
    pred_label = np.argmax(pred[0])
    pred_conf = pred[0][pred_label] * 100

    try:
        # Generate Grad-CAM
        heatmap, _ = compute_gradcam_simple(CNN_2D_best_model, img, pred_label, last_conv_layer)
        overlay, heatmap_resized = create_overlay(heatmap, img.squeeze())

        # Plot
        axes[i, 0].imshow(img.squeeze(), cmap='gray')
        axes[i, 0].set_title(f'Input Signal\nTrue: Class {true_label}', fontsize=10)
        axes[i, 0].axis('off')

        axes[i, 1].imshow(heatmap_resized, cmap='jet')
        axes[i, 1].set_title(f'Grad-CAM Heatmap\nPred: Class {pred_label} ({pred_conf:.1f}%)', fontsize=10)
        axes[i, 1].axis('off')

        axes[i, 2].imshow(overlay)
        axes[i, 2].set_title('Overlay', fontsize=10)
        axes[i, 2].axis('off')

        success_count += 1
        print(f"✓ Sample {i+1}/{num_samples}: True={true_label}, Pred={pred_label} ({pred_conf:.1f}%)")

    except Exception as e:
        print(f"✗ Sample {i+1}/{num_samples}: Error - {str(e)[:100]}")
        import traceback
        traceback.print_exc()

        axes[i, 0].imshow(img.squeeze(), cmap='gray')
        axes[i, 0].set_title(f'Input\nTrue: {true_label}', fontsize=10)
        axes[i, 0].axis('off')

        axes[i, 1].text(0.5, 0.5, 'Error', ha='center', va='center')
        axes[i, 1].axis('off')

        axes[i, 2].text(0.5, 0.5, 'Error', ha='center', va='center')
        axes[i, 2].axis('off')

plt.tight_layout()
output_file = f'{gradcam_dir}gradcam_visualization_random.png'
plt.savefig(output_file, dpi=150, bbox_inches='tight')
plt.show()

print(f"\n✅ Saved: {output_file}")
print(f"✅ Success rate: {success_count}/{num_samples}")

# ============================================================================
# Generate class-wise
# ============================================================================

print("\n" + "=" * 70)
print("GENERATING CLASS-WISE GRAD-CAM")
print("=" * 70)

num_classes = 10
fig, axes = plt.subplots(num_classes, 3, figsize=(12, 4*num_classes))

print(f"\nGenerating Grad-CAM for all {num_classes} classes...\n")

class_success = 0
for class_idx in range(num_classes):
    class_samples = [i for i in range(len(y_2D_test))
                     if np.argmax(y_2D_test[i]) == class_idx]

    if len(class_samples) == 0:
        print(f"⚠ Class {class_idx}: No samples")
        axes[class_idx, 0].text(0.5, 0.5, f'No samples\nClass {class_idx}',
                                ha='center', va='center')
        axes[class_idx, 0].axis('off')
        axes[class_idx, 1].axis('off')
        axes[class_idx, 2].axis('off')
        continue

    idx = class_samples[0]
    img = X_2D_test[idx]

    pred = CNN_2D_best_model.predict(img[np.newaxis, ...], verbose=0)
    pred_label = np.argmax(pred[0])
    pred_conf = pred[0][pred_label] * 100

    try:
        heatmap, _ = compute_gradcam_simple(CNN_2D_best_model, img, class_idx, last_conv_layer)
        overlay, heatmap_resized = create_overlay(heatmap, img.squeeze())

        axes[class_idx, 0].imshow(img.squeeze(), cmap='gray')
        axes[class_idx, 0].set_title(f'Input Signal\nClass {class_idx}', fontsize=10)
        axes[class_idx, 0].axis('off')

        axes[class_idx, 1].imshow(heatmap_resized, cmap='jet')
        axes[class_idx, 1].set_title(f'Grad-CAM\nPred: {pred_label} ({pred_conf:.1f}%)', fontsize=10)
        axes[class_idx, 1].axis('off')

        axes[class_idx, 2].imshow(overlay)
        axes[class_idx, 2].set_title('Overlay', fontsize=10)
        axes[class_idx, 2].axis('off')

        class_success += 1
        correct = "✓" if pred_label == class_idx else "✗"
        print(f"{correct} Class {class_idx}: Predicted as {pred_label} ({pred_conf:.1f}%)")

    except Exception as e:
        print(f"✗ Class {class_idx}: Error - {str(e)[:100]}")

        axes[class_idx, 0].imshow(img.squeeze(), cmap='gray')
        axes[class_idx, 0].set_title(f'Class {class_idx}', fontsize=10)
        axes[class_idx, 0].axis('off')

        axes[class_idx, 1].text(0.5, 0.5, 'Error', ha='center', va='center')
        axes[class_idx, 1].axis('off')

        axes[class_idx, 2].text(0.5, 0.5, 'Error', ha='center', va='center')
        axes[class_idx, 2].axis('off')

plt.tight_layout()
output_file = f'{gradcam_dir}gradcam_visualization_all_classes.png'
plt.savefig(output_file, dpi=150, bbox_inches='tight')
plt.show()

print(f"\n✅ Saved: {output_file}")
print(f"✅ Success rate: {class_success}/{num_classes}")

print("\n" + "=" * 70)
print("DONE!")
print("=" * 70)
```

```python
def compute_sparsity(heatmap, threshold=0.5):
    """
    Measures how focused the explanation is
    Lower = more focused (better)
    """
    active_pixels = np.sum(heatmap > threshold)
    total_pixels = heatmap.size
    return active_pixels / total_pixels
```
